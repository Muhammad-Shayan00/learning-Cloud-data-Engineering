# Introduction to numpy

```python
import numpy as np
```

```python
print(np.__version__)
```

```python
# to create a numpy array
myarr=np.array([12,14,15],np.int64)#np.int64 for memory management
print(myarr)
```

```python
# to create a 2d numpy array
myarr2=np.array([[12,14,15],[1,2,3,]])
print(myarr2)
print(np.shape(myarr2))#tell numbers of rows and columns
print(myarr2[0,1])#to print a letter
myarr2[1,1]=50
print(myarr2)
```

```python
zeross=np.zeros((2,5))
print(zeross)
rng=np.arange((15))
print(rng)
lspace=np.linspace(1,15,10)#equally space 10 element from 1 to 15 
print(lspace)
identity_array=np.identity(4)#make identity array
print(identity_array)
```

```python
arr3=np.arange(99)
print(arr3)
arr4=arr3.reshape(3,33)#//to split in 3 of 33
print(arr4)
arr5=arr3.ravel()
print(arr5)#to convert in 1d array

```

```python
x=[[1,2,3],[4,5,6],[6,7,8]]
arr6=np.array(x)
print(arr6)
print(arr6.sum(axis=0))#sum all y axis element
print(arr6.sum(axis=1))#sum all x axis element
```

```python
#to generate random number
arr1=np.random.random(5)
print(arr1)
#for specific range
arr2=np.random.uniform(0.1,5.0,5)
print(arr2)


```
