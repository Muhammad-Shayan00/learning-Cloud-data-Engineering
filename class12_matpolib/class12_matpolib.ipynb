## basic plot of Matploatlib

```python
import matplotlib.pyplot as plt

```

```python
year=[2005,2006,2008,2010]
population=[6,8,30,40]
plt.plot(year,population)
plt.xlabel("years")
plt.ylabel("population")
plt.title("population vs years")
plt.show()


```

```python
plt.bar(year,population)
plt.xlabel("years")
plt.ylabel("population")
plt.title("population vs years")
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y, marker='o')
plt.yticks([10, 20, 30, 40], ["Low", "Medium", "High", "Very High"])#change number to  text
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Custom Y-ticks Example")
plt.show()
```
