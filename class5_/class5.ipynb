# LOOPS

```python
for i in range (6):
    print("happy birthday")
```

```python
n=int(input("enter the number"))
for i in range(1,n+1):
    print(f"the square of {i} is {i*i}")
```

```python
table=int(input("enter the table:"))
number=int(input("enter the numbers of times:"))

for i in range (1,number+1):
    print(f"{table} x {i} = {i*table}")
```

```python
n=int(input("enter the number of times"))
n=n+1
for i in range(2,6):
    n=n-1
    for j in range(1,n+1):
        print(f"{i} x {j} = {i*j}")
    print(" ")    
    print("next table")    
```

# Dictionary

```python
bio_data={"name":"shayan",
          "f_name":"Mushtaq",
          "age":20,

}
print(bio_data)
print(bio_data["name"])

#update value
bio_data["age"]=21
print(bio_data["age"])

#add value
bio_data["is_present"]=True
print(bio_data["is_present"])

#remove items from dictionary
bio_data.pop("f_name")
print(bio_data)

#using loop

for key,value in bio_data.items():
    print(f"the {key} is {value}")
```
